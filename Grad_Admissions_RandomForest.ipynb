## Regression Model 

```python
y = grad_data['Chance of Admit ']
X = grad_data.drop(columns='Chance of Admit ')
X = X.drop(columns='Serial No.')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size= .90, random_state=2)
```

```python
print(X.head())
X.columns
```

```python
print(y.head())
```

```python
grad_model = RandomForestRegressor(random_state=2, n_estimators=500)
grad_model.fit(X_train, y_train)
```

```python
predictions = grad_model.predict(X_test)
grad_mae = mean_absolute_error(predictions, y_test)
print(str(grad_mae * 100) + "%")
```
